# Data Transformations

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

## Long Form DataFrame

```python
groups = np.repeat(['A', 'B', 'C'], 10)
```

```python
features = np.tile(['f1', 'f2', 'f3'], 10)[:30]
```

```python
values = np.random.randint(1, 100, size=30).astype(float)
```

```python
# # Introduce some missing values
# values[[3, 7, 15]] = np.nan
```

```python
long_df = pd.DataFrame({
    'group': groups,
    'feature': features,
    'value': values
})
```

```python
long_df
```

## Wide Form DataFrame

```python
wide_groups = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
```

```python
wide_data = {
    'id': range(1, 11),
    'group': wide_groups,
    'f1': np.random.randint(1, 50, 10).astype(float),
    'f2': np.random.randint(50, 100, 10).astype(float),
    'f3': np.random.randint(100, 150, 10).astype(float)
}
```

```python
wide_df = pd.DataFrame(wide_data)
```
